Clustering analsyis

```python
from sklearn.cluster import KMeans

#removing  the labels column
x=data.drop(['label'],axis=1)
x=x.values

print(x.shape)
```

```python
## determining the optimumm  number of clusters within the dataset
plt.rcParams['figure.figsize']=(10,4)

wccs= []
for i in range(1,11):
  km=KMeans(n_clusters=i ,init='k-means++',max_iter=300, n_init=10,random_state=0)
  km.fit(x)
  wccs.append(km.inertia_)

# lets plot  the results

plt.plot(range(1,11),wccs)
plt.title('The ELbow  method',fontsize=20)
plt.xlabel('No.of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means = km.fit(x)

a = data['label']
z = pd.concat([pd.DataFrame(y_means.labels_, columns=['cluster']), a], axis=1)
z = z.rename(columns={'label': 'crop'})

print("Crops in each cluster:\n")
print("Crops in cluster 0:", z[z['cluster'] == 0]['crop'].unique())
print('-------------------------------------------------------------')
print("Crops in cluster 1:", z[z['cluster'] == 1]['crop'].unique())
print('-------------------------------------------------------------')
print("Crops in cluster 2:", z[z['cluster'] == 2]['crop'].unique())
print('-------------------------------------------------------------')
print("Crops in cluster 3:", z[z['cluster'] == 3]['crop'].unique())
print('-------------------------------------------------------------')
```

```python
## predicitive modelling
y=data['label']
x=data.drop(['label'],axis=1)

print("shape of x",x.shape)
print("shape of y",y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)

print("shape of x_train",x_train.shape)
print("shape of y_train",y_train.shape)
print("shape of x_test",x_test.shape)
print("shape of y_test",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(solver='saga',max_iter=3000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix


## printing the confusion matrix first

plt.rcParams['figure.figsize']=(10,10)
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,cmap='Wistia')
plt.title('confusion matrix for logistic regreesion',fontsize=15)
plt.show()
```

```python
from sklearn.metrics import classification_report

cr=classification_report(y_test,y_pred)
print(cr)
```
